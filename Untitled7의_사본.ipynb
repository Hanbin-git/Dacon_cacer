```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import f1_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Í≤ΩÎ°ú
def get_path(filename):
    return "/content/open_1/" + filename

# Îç∞Ïù¥ÌÑ∞ Î°úÎî©
train = pd.read_csv(get_path("train.csv"))
test = pd.read_csv(get_path("test.csv"))
submission = pd.read_csv(get_path("sample_submission.csv"))

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# ‚úÖ ÌååÏÉù Î≥ÄÏàò ÏÉùÏÑ± Ìï®Ïàò
def add_derived_features(df):
    df = df.copy()
    for col in df.select_dtypes(include='number'):
        df[f"{col}_squared"] = df[col] ** 2
        df[f"{col}_sqrt"] = np.sqrt(np.abs(df[col]))
        df[f"{col}_log"] = np.log1p(np.abs(df[col]))

    # ÎèÑÎ©îÏù∏ ÏßÄÏãù Í∏∞Î∞ò ÌååÏÉùÎ≥ÄÏàò
    if "T4_Result" in df.columns and "TSH_Result" in df.columns:
        df["T4_TSH_ratio"] = df["T4_Result"] / (df["TSH_Result"] + 1e-3)
    if "T3_Result" in df.columns and "Nodule_Size" in df.columns:
        df["T3_times_Nodule"] = df["T3_Result"] * df["Nodule_Size"]
    if "Weight_Risk" in df.columns and "Age" in df.columns:
        df["Age_Weight_risk"] = df["Age"] * df["Weight_Risk"]
    return df

X = add_derived_features(X)
X_test = add_derived_features(X_test)

# ‚úÖ Ï†ÑÏ≤òÎ¶¨
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# ‚úÖ Î™®Îç∏ Î¶¨Ïä§Ìä∏ Ï†ïÏùò
models = {
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# ‚úÖ Í∞Å Î™®Îç∏Î≥Ñ ÏÑ±Îä• ÎπÑÍµê Î∞è Ï†úÏ∂ú ÌååÏùº ÏÉùÏÑ±
for name, model in models.items():
    print(f"\nüîç Î™®Îç∏: {name}")

    oof_preds = cross_val_predict(model, X, y, method="predict_proba", cv=5)[:, 1]
    thresholds = np.arange(0.1, 0.9, 0.01)
    f1s = [f1_score(y, oof_preds > t) for t in thresholds]
    best_thresh = thresholds[np.argmax(f1s)]
    print(f"‚úÖ Best F1: {np.max(f1s):.4f}, Threshold: {best_thresh:.2f}")

    # Ï†ÑÏ≤¥ ÌïôÏäµ ÌõÑ ÏòàÏ∏°
    model.fit(X, y)
    test_preds = model.predict_proba(X_test)[:, 1]

    # Ï†ÄÏû•
    sub = submission.copy()
    sub["Cancer"] = (test_preds > best_thresh).astype(int)
    sub.to_csv(f"submission_{name.lower()}.csv", index=False)
```

```python
import optuna
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import f1_score, make_scorer

# F1 Í∏∞Î∞ò CV ÌèâÍ∞Ä Ìï®Ïàò
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, 0.3, log=True),
        "max_depth": trial.suggest_int("max_depth", 3, 12),
        "num_leaves": trial.suggest_int("num_leaves", 15, 100),
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 50),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "random_state": 42,
        "n_jobs": -1
    }

    model = LGBMClassifier(**params)
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    score = cross_val_score(model, X, y, cv=skf, scoring=make_scorer(f1_score)).mean()
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30)

# ÏµúÏ†Å Î™®Îç∏ ÌõàÎ†®
best_params = study.best_trial.params
model = LGBMClassifier(**best_params)
model.fit(X, y)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import f1_score
from lightgbm import LGBMClassifier
import os

# Í≤ΩÎ°ú
def get_path(filename):
    return "/content/open_1/" + filename

# Îç∞Ïù¥ÌÑ∞ Î°úÎî©
train = pd.read_csv(get_path("train.csv"))
test = pd.read_csv(get_path("test.csv"))
submission = pd.read_csv(get_path("sample_submission.csv"))

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# ‚úÖ ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ±
def add_derived_features(df):
    df = df.copy()
    df["T4_TSH_ratio"] = df["T4_Result"] / (df["TSH_Result"] + 1e-3)
    df["T3_times_Nodule"] = df["T3_Result"] * df["Nodule_Size"]
    return df

X = add_derived_features(X)
X_test = add_derived_features(X_test)

# ‚úÖ Ï†ÑÏ≤òÎ¶¨
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# ‚úÖ Ï§ëÏöîÎèÑ ÎÇÆÏùÄ Î≥ÄÏàò Ï†úÍ±∞ (ÏÇ¨Ï†Ñ Î∂ÑÏÑù Í≤∞Í≥º Í∏∞Î∞ò)
low_importance_features = [
    "Age_squared", "Age_sqrt", "Age_log", "Nodule_Size_log",
]
X = X.drop(columns=[col for col in low_importance_features if col in X.columns])
X_test = X_test.drop(columns=[col for col in low_importance_features if col in X_test.columns])

# ‚úÖ Î™®Îç∏ ÌïôÏäµ (5-Fold)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
test_preds = np.zeros(len(X_test))
oof_preds = np.zeros(len(X))

for train_idx, val_idx in skf.split(X, y):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = LGBMClassifier(random_state=42)
    model.fit(X_train, y_train)

    oof_preds[val_idx] = model.predict_proba(X_val)[:, 1]
    test_preds += model.predict_proba(X_test)[:, 1] / skf.n_splits

# ‚úÖ Threshold ÏµúÏ†ÅÌôî
thresholds = np.arange(0.1, 0.9, 0.01)
f1s = [f1_score(y, oof_preds > t) for t in thresholds]
best_thresh = thresholds[np.argmax(f1s)]

print(f"Best Threshold: {best_thresh:.2f} | Best F1 Score: {np.max(f1s):.4f}")

# ‚úÖ Ï†úÏ∂ú ÌååÏùº Ï†ÄÏû•
submission["Cancer"] = (test_preds > best_thresh).astype(int)
submission.to_csv("submission_lgbm_pruned.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import shap
from lightgbm import LGBMClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.metrics import f1_score

# Í≤ΩÎ°ú Ìï®Ïàò
def get_path(filename):
    return "/content/open_1/" + filename

# Îç∞Ïù¥ÌÑ∞ Î°úÎî©
train = pd.read_csv(get_path("train.csv"))
test = pd.read_csv(get_path("test.csv"))
submission = pd.read_csv(get_path("sample_submission.csv"))

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# Ï†ÑÏ≤òÎ¶¨
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# Î™®Îç∏ ÌïôÏäµ
model = LGBMClassifier(random_state=42)
model.fit(X, y)

# SHAP ÏÉÅÌò∏ÏûëÏö© Í∞í Í≥ÑÏÇ∞
explainer = shap.TreeExplainer(model)
interaction_values = explainer.shap_interaction_values(X)  # (N, F, F)

# ÏÉÅÌò∏ÏûëÏö© ÌèâÍ∑†Í∞í Í≥ÑÏÇ∞ ‚Üí (F, F)
interaction_mean = np.mean(interaction_values, axis=0)

# DataFrame Î≥ÄÌôò Î∞è ÏÉÅÏúÑ ÏÉÅÌò∏ÏûëÏö© Ï∂îÏ∂ú
interaction_df = pd.DataFrame(interaction_mean, columns=X.columns, index=X.columns)
upper_triangle = interaction_df.where(np.triu(np.ones(interaction_df.shape), k=1).astype(bool))
top_interactions = upper_triangle.stack().sort_values(ascending=False).head(5)

# ÌååÏÉù Î≥ÄÏàò ÏÉùÏÑ±
for f1, f2 in top_interactions.index:
    new_col = f"{f1}_x_{f2}"
    X[new_col] = X[f1] * X[f2]
    X_test[new_col] = X_test[f1] * X_test[f2]

# ÏµúÏ¢Ö Î™®Îç∏ ÌïôÏäµ
model_final = LGBMClassifier(random_state=42)
model_final.fit(X, y)

# ÌÖåÏä§Ìä∏ÏÖã ÏòàÏ∏°
test_preds = model_final.predict_proba(X_test)[:, 1]

# OOF Í∏∞Î∞ò Threshold ÌäúÎãù
oof_preds = cross_val_predict(model_final, X, y, method="predict_proba", cv=5)[:, 1]
thresholds = np.arange(0.1, 0.9, 0.01)
f1s = [f1_score(y, oof_preds > t) for t in thresholds]
best_thresh = thresholds[np.argmax(f1s)]
print(f"Best F1 = {max(f1s):.4f} at Threshold = {best_thresh:.2f}")

# Ï†úÏ∂ú ÌååÏùº ÏÉùÏÑ±
submission["Cancer"] = (test_preds > best_thresh).astype(int)
submission.to_csv("submission_lgbm_shap_interaction.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import f1_score
from lightgbm import LGBMClassifier
from itertools import combinations
from tqdm import tqdm

# üî∏ Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
train = pd.read_csv("/content/open_1/train.csv")
test = pd.read_csv("/content/open_1/test.csv")
submission = pd.read_csv("/content/open_1/sample_submission.csv")

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# üî∏ Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# üî∏ ÌõÑÎ≥¥ ÍµêÌò∏ÏûëÏö© Î≥ÄÏàò Ï†ïÏùò
interaction_pairs = [
    ("Age", "Nodule_Size"),
    ("T3_Result", "T4_Result"),
    ("Weight_Risk", "Iodine_Deficiency"),
    ("Race", "Country"),
    ("TSH_Result", "T4_Result"),
    ("Gender", "Family_Background")
]

# üî∏ ÌèâÍ∞Ä Ìï®Ïàò
def evaluate_interaction_combination(pair_subset):
    X_aug = X.copy()
    for f1, f2 in pair_subset:
        col_name = f"{f1}_x_{f2}"
        X_aug[col_name] = X[f1] * X[f2]

    model = LGBMClassifier(random_state=42)
    oof_preds = cross_val_predict(model, X_aug, y, method="predict_proba", cv=5)[:, 1]

    thresholds = np.arange(0.1, 0.9, 0.01)
    f1s = [f1_score(y, oof_preds > t) for t in thresholds]
    best_f1 = max(f1s)
    best_thresh = thresholds[np.argmax(f1s)]

    return best_f1, best_thresh, pair_subset

# üî∏ Ï°∞Ìï©Î≥Ñ ÌèâÍ∞Ä (3~6Í∞úÏî©)
results = []
for k in range(3, 7):
    for comb in tqdm(list(combinations(interaction_pairs, k)), desc=f"{k}-way interactions"):
        score, threshold, used_pairs = evaluate_interaction_combination(comb)
        results.append({"F1": score, "Threshold": threshold, "Pairs": used_pairs})

# üî∏ Í≤∞Í≥º Ï†ïÎ†¨
results_df = pd.DataFrame(results).sort_values(by="F1", ascending=False)
print("Top 5 Combinations:")
print(results_df.head(5))
```

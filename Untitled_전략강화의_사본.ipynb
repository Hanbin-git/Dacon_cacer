```python
# ì „ëžµ: íŒŒìƒë³€ìˆ˜ ì œì™¸ + CV ì•™ìƒë¸”(TTA ëŒ€ì²´ìš©)
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import f1_score
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# âœ… ê²½ë¡œ í•¨ìˆ˜
def get_path(filename):
    return "/content/open_1/" + filename  # í™˜ê²½ì— ë§žê²Œ ê²½ë¡œ ìˆ˜ì •

# âœ… ë°ì´í„° ë¡œë”©
train = pd.read_csv(get_path("train.csv"))
test = pd.read_csv(get_path("test.csv"))
submission = pd.read_csv(get_path("sample_submission.csv"))

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# âœ… ì „ì²˜ë¦¬ (Label Encoding + Imputation)
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# âœ… ëª¨ë¸ ì •ì˜
model_lgbm = LGBMClassifier(learning_rate=0.05, max_depth=6, num_leaves=31, random_state=42)
model_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model_cat = CatBoostClassifier(verbose=0, random_state=42)
meta_model = LGBMClassifier(learning_rate=0.05, max_depth=3, num_leaves=15, random_state=42)

# âœ… CV ê¸°ë°˜ ì•™ìƒë¸” ì˜ˆì¸¡
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
test_preds_total = np.zeros(len(X_test))
oof_preds = np.zeros(len(X))
thresholds = np.arange(0.1, 0.9, 0.01)
f1_scores_list = []

for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):
    print(f"\nðŸš€ Fold {fold+1}/{n_splits}")
    X_tr, X_val = X.iloc[train_idx], X.iloc[valid_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[valid_idx]

    # âœ… ìŠ¤íƒœí‚¹ ëª¨ë¸ ì •ì˜
    base_models = [
        ('lgbm', model_lgbm),
        ('xgb', model_xgb),
        ('cat', model_cat)
    ]

    stack_model = StackingClassifier(
        estimators=base_models,
        final_estimator=meta_model,
        cv=5,
        stack_method="predict_proba",
        n_jobs=-1
    )

    stack_model.fit(X_tr, y_tr)
    val_preds = stack_model.predict_proba(X_val)[:, 1]
    test_fold_preds = stack_model.predict_proba(X_test)[:, 1]

    oof_preds[valid_idx] = val_preds
    test_preds_total += test_fold_preds / n_splits  # CV ì•™ìƒë¸”

    # âœ… Foldë³„ Best Threshold íƒìƒ‰
    fold_f1s = [f1_score(y_val, val_preds > t) for t in thresholds]
    best_t = thresholds[np.argmax(fold_f1s)]
    best_f1 = max(fold_f1s)
    print(f"ðŸ“ˆ Fold {fold+1} Best Threshold: {best_t:.2f}, F1 Score: {best_f1:.5f}")
    f1_scores_list.append(best_f1)

# âœ… ì „ì²´ Best Threshold (ì „ì²´ OOF ê¸°ì¤€)
overall_f1s = [f1_score(y, oof_preds > t) for t in thresholds]
best_thresh = thresholds[np.argmax(overall_f1s)]
print(f"\nâœ… Overall Best Threshold: {best_thresh:.2f}, Overall F1 Score: {max(overall_f1s):.5f}")
print(f"ðŸ“Š Fold F1 Mean: {np.mean(f1_scores_list):.5f}, Std: {np.std(f1_scores_list):.5f}")

# âœ… ìµœì¢… ì˜ˆì¸¡ ë° ì €ìž¥
submission["Cancer"] = (test_preds_total > best_thresh).astype(int)
submission.to_csv("submission_stacking_cv_ensemble.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import f1_score
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# ë°ì´í„° ë¡œë“œ í•¨ìˆ˜
def get_path(filename):
    return "/content/open_1/" + filename

train = pd.read_csv(get_path("train.csv"))
test = pd.read_csv(get_path("test.csv"))
submission = pd.read_csv(get_path("sample_submission.csv"))

# íš¨ê³¼ ìžˆëŠ” íŒŒìƒë³€ìˆ˜ ì¶”ê°€
def create_features(df):
    df = df.copy()
    df["Age_x_Nodule_Size"] = df["Age"] * df["Nodule_Size"]
    df["T3_x_Nodule_Size"] = df["T3_Result"] * df["Nodule_Size"]
    df["Age_div_Nodule_Size"] = df["Age"] / (df["Nodule_Size"] + 1e-5)
    df["Age_category"] = pd.cut(df["Age"], bins=[0, 30, 45, 60, 100], labels=[0, 1, 2, 3])
    df["Nodule_Category"] = pd.cut(df["Nodule_Size"], bins=[-0.1, 5, 10, 20, 100], labels=[0, 1, 2, 3])
    return df

train = create_features(train)
test = create_features(test)

y = train["Cancer"]
X = train.drop(columns=["ID", "Cancer"])
X_test = test.drop(columns=["ID"])

# ì „ì²˜ë¦¬ (ë¼ë²¨ ì¸ì½”ë”© + ê²°ì¸¡ì¹˜ ì²˜ë¦¬)
def preprocess(df, encoders=None):
    df = df.copy()
    encoders = encoders or {}
    for col in df.columns:
        if df[col].dtype == "object" or df[col].dtype.name == "category":
            df[col] = df[col].astype(str)
            if col not in encoders:
                encoders[col] = LabelEncoder().fit(df[col])
            df[col] = encoders[col].transform(df[col])
        else:
            df[col] = SimpleImputer(strategy="mean").fit_transform(df[[col]])
    return df, encoders

X, encoders = preprocess(X)
X_test, _ = preprocess(X_test, encoders)

# ëª¨ë¸ ì •ì˜
model_lgbm = LGBMClassifier(learning_rate=0.05, max_depth=6, num_leaves=31, random_state=42)
model_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model_cat = CatBoostClassifier(verbose=0, random_state=42)
meta_model = LGBMClassifier(learning_rate=0.05, max_depth=3, num_leaves=15, random_state=42)

base_models = [
    ('lgbm', model_lgbm),
    ('xgb', model_xgb),
    ('cat', model_cat)
]

stack_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=5,
    stack_method="predict_proba",
    n_jobs=-1
)

# OOF ì˜ˆì¸¡ ë° ìµœì  Threshold ê³„ì‚°
oof_preds = cross_val_predict(stack_model, X, y, method="predict_proba", cv=5)[:, 1]
thresholds = np.arange(0.1, 0.9, 0.01)
f1s = [f1_score(y, oof_preds > t) for t in thresholds]
best_thresh = thresholds[np.argmax(f1s)]

# í•™ìŠµ ë° ì˜ˆì¸¡
stack_model.fit(X, y)
test_preds = stack_model.predict_proba(X_test)[:, 1]
submission["Cancer"] = (test_preds > best_thresh).astype(int)
submission.to_csv("submission_stacking_selected_features.csv", index=False)
```
